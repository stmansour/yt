```python
import json
import pandas as pd
import matplotlib

file = "steve_mansour.json"
data = None
with open(file,'r') as f:
    data = json.load(f)
    
channel_id, stats = data.popitem()
channel_stats = stats["channel_statistics"]
video_stats = stats["video_data"]
print( f" View Count: {channel_stats['viewCount']}")
print( f"     Videos: {channel_stats['videoCount']}")
print( f"Subscribers: {channel_stats['subscriberCount']}")
```

```python
# video statistics
sorted_vids = sorted(video_stats.items(),key=lambda item:int(item[1]["viewCount"]), reverse=False)
stats = []
for vid in sorted_vids:
    video_id = vid[0]
    title = vid[1]["title"]
    likes = int(vid[1]["likeCount"])
    comments = int(vid[1]["commentCount"])
    views = int(vid[1]["viewCount"])
    date = vid[1]["publishedAt"]
    stats.append([date,title,views,likes,comments])
```

```python
df = pd.DataFrame(stats, columns=["date","title", "views", "likes", "comments"])
df
```

```python
top10 = df.head(10)
ax = top10.plot.bar(x="title", y="views", figsize=(12,8))
```
